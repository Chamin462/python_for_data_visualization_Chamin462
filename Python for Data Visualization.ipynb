# Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y= 2*x
y
```

```python
plt.plot(x,y, color='g',linestyle=':',linewidth=4)
plt.title("Line Plot")
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.show()
```

```python
# Adding two line in the same plot.
y1= 2*x
y2= 3*x
```

```python
plt.plot(x,y1, color='g',linestyle=':',linewidth=4)
plt.plot(x,y2, color='r',linestyle=':',linewidth=4)
plt.title("Line Plot")
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.grid(True)
plt.show()
```

```python
# Adding sub-plot method (1,2,1) show 1 show 1rows and 2 show 2columns and 1 show first subplot index ha.
plt.subplot(1,2,1)
plt.plot(x,y1, color='g',linestyle=':',linewidth=4)
plt.subplot(1,2,2)
plt.plot(x,y2, color='r',linestyle=':',linewidth=4)
plt.title("Line Plot")
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.grid(True)
plt.show()
```

# Bar plot

# Scatter Plot
creating a basic scatter_plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,6,5,7,4,3,2,9,1]
plt.scatter(x,a, marker="*",c='r',s=20)
plt.show()
```

```python
# Adding two markers in the same plot
x=[10,20,30,40,50,60,70,80,90]
a=[8,6,5,7,4,3,2,9,1]
b=[5,4,6,3,2,1,7,8,9]
plt.scatter(x,a, marker="*",c='r',s=50)
plt.scatter(x,b, marker=".",c='g',s=50)

plt.show()
```

```python
# Adding sub-plot
x=[10,20,30,40,50,60,70,80,90]
a=[8,6,5,7,4,3,2,9,1]
b=[5,4,6,3,2,1,7,8,9]
plt.subplot(1,2,1)
plt.scatter(x,a, marker="*",c='r',s=50)
plt.subplot(1,2,2)
plt.scatter(x,b, marker=".",c='g',s=50)

plt.show()
```
